# Liquid Neural Networks (LNN)
En este notebook exploraremos el concepto de las Liquid Neural Networks (LNNs), su inspiración biológica, su arquitectura y sus principales características. Este documento está enfocado en proporcionar una visión teórica sobre las LNNs, que son redes neuronales capaces de adaptarse a los cambios en tiempo real mediante el uso de ecuaciones diferenciales ordinarias (ODEs).

## Introducción
Las Liquid Neural Networks (LNNs) representan un nuevo enfoque en el diseño de redes neuronales, inspiradas en sistemas biológicos como el sistema nervioso del nematodo *C. elegans*. A diferencia de las redes neuronales tradicionales, las LNNs no dependen de pesos estáticos o funciones de activación fijas. En su lugar, utilizan ecuaciones diferenciales para modelar el comportamiento dinámico de las neuronas, lo que les permite adaptarse de forma continua a los cambios en los datos de entrada.


## Inspiración Biológica
Las LNNs están inspiradas en el sistema nervioso de organismos biológicos simples como el *C. elegans*, un nematodo con solo 302 neuronas, capaz de generar comportamientos complejos. Este sistema biológico sirve como modelo para las LNNs debido a su capacidad para adaptarse a cambios en el entorno con una arquitectura neuronal limitada.

El sistema nervioso del *C. elegans* se caracteriza por conexiones neuronales dinámicas que permiten la transmisión de señales entre las neuronas de manera continua, lo que resulta en una adaptación eficiente a los cambios externos. Esta capacidad de adaptación dinámica es una de las principales características que las LNNs buscan replicar.


## Definición de las Liquid Neural Networks (LNNs)
Una Liquid Neural Network (LNN) es una red neuronal que utiliza ecuaciones diferenciales para modelar el comportamiento de sus neuronas de manera continua en el tiempo. En lugar de tener pesos y funciones de activación fijas como en las redes neuronales convencionales, las LNNs pueden modificar sus conexiones internas y la fuerza de esas conexiones en función de las entradas.

### Ecuación Diferencial de las LNNs
La dinámica de las LNNs puede modelarse utilizando ecuaciones diferenciales de primer orden, lo que permite que el estado de las neuronas cambie en función del tiempo. La ecuación básica que describe el comportamiento de una LNN es:

$$
\frac{dx(t)}{dt} = f(x(t), I(t), t, \theta)
$$

Donde:
- $ x(t) $ es el estado de la neurona en el tiempo $ t $.
- $ I(t) $ es la entrada que recibe la neurona en el tiempo $ t $.
- $ \theta $ son los parámetros de la red (incluyen los pesos y sesgos).
- $ f $ es una función no lineal que describe la evolución del estado.


## Arquitectura de las Liquid Neural Networks
La arquitectura de las LNNs difiere significativamente de las redes neuronales tradicionales. En lugar de una estructura de capas completamente conectadas con pesos estáticos, las LNNs tienen una arquitectura que puede cambiar dinámicamente. Algunas de las características principales incluyen:

1. **Conexiones Dinámicas**: Las LNNs pueden ajustar las conexiones entre sus neuronas en función de las entradas, lo que les permite aprender y adaptarse de manera continua.
   
2. **Ecuaciones Diferenciales**: La evolución del estado de cada neurona está gobernada por una ecuación diferencial ordinaria (ODE), lo que les permite actualizar su estado de manera continua en el tiempo.
   
3. **Constante de Tiempo Líquida (Liquid Time Constant, LTC)**: Un parámetro clave de las LNNs es la constante de tiempo líquida, que controla la velocidad con la que el estado de la neurona responde a las entradas. Este parámetro es adaptable, lo que permite una mayor flexibilidad en el comportamiento de la red.

### Ecuación con LTC
Una LNN típica introduce la constante de tiempo líquida \( \tau \) en su ecuación diferencial, lo que la convierte en:

$$
\frac{dx(t)}{dt} = \frac{-x(t) + f(x(t), I(t), \theta)}{\tau}
$$

Donde $ \tau $ es la constante de tiempo líquida, que define la rapidez con la que los estados cambian. Este parámetro permite que las LNNs tengan un control más granular sobre su adaptación a las nuevas entradas.


## Características Principales de las LNNs

1. **Aprendizaje Continuo**: A diferencia de las redes tradicionales, que requieren un entrenamiento completo para ajustar sus pesos, las LNNs pueden aprender y adaptarse de manera continua. Esto les permite ajustarse a cambios en los datos de entrada en tiempo real.

2. **Arquitectura Dinámica**: Las LNNs pueden modificar dinámicamente el número de neuronas y las conexiones entre ellas. Esto significa que pueden aumentar o reducir su complejidad según las necesidades de la tarea y los datos que están procesando.

3. **Procesamiento en Tiempo Real**: Gracias a su capacidad de cambiar su estado de forma continua, las LNNs son adecuadas para aplicaciones que requieren procesamiento en tiempo real, como el análisis de series temporales o el procesamiento de señales.

4. **Robustez a Ruido**: Las LNNs son más resilientes frente a entradas ruidosas o perturbaciones, lo que las hace adecuadas para aplicaciones donde las señales de entrada son inciertas o contienen ruido.


## Ventajas de las LNNs sobre las Redes Neuronales Tradicionales

1. **Menor necesidad de datos etiquetados**: Las LNNs no requieren grandes cantidades de datos etiquetados para entrenar. Esto las hace ideales para aplicaciones donde el etiquetado manual de datos es costoso o no es práctico.

2. **Eficiencia computacional**: Las LNNs tienen un tamaño de modelo más pequeño y requieren menos recursos de computación para procesar datos, lo que las hace más escalables y aplicables en entornos empresariales.

3. **Mejor manejo de series temporales**: Debido a su naturaleza continua, las LNNs pueden manejar de manera más eficiente los datos secuenciales y las series temporales, donde los patrones varían con el tiempo.

4. **Adaptabilidad**: Las LNNs pueden adaptarse rápidamente a cambios en los datos de entrada, lo que las hace adecuadas para aplicaciones en tiempo real donde los datos cambian constantemente, como el seguimiento de objetos, predicción de series temporales, y procesamiento de lenguaje natural.


## Casos de Uso de las LNNs

1. **Procesamiento de series temporales**: Las LNNs son muy adecuadas para procesar datos secuenciales, como precios de acciones, datos climáticos o señales biomédicas. Su capacidad para adaptarse a los cambios en tiempo real las hace especialmente útiles en estas áreas.

2. **Procesamiento de imágenes y videos**: Las LNNs se pueden utilizar en tareas de procesamiento de visión por computadora, como el seguimiento de objetos, la segmentación de imágenes y el reconocimiento de patrones, debido a su capacidad para adaptarse a patrones temporales complejos.

3. **Comprensión del lenguaje natural**: Las LNNs pueden ser aplicadas en tareas de procesamiento del lenguaje natural, como el análisis de sentimientos y la traducción automática, debido a su capacidad para aprender de datos en tiempo real y adaptarse a cambios en los patrones de lenguaje.

4. **Sistemas de control en tiempo real**: Las LNNs pueden ser usadas para controlar sistemas autónomos como robots, drones y vehículos autónomos, gracias a su capacidad para adaptarse rápidamente a cambios en el entorno.

### Arquitectura de la LNN:


Vamos a definir la clase que representará nuestra Liquid Neural Network. En este ejemplo, utilizaremos PyTorch para crear una estructura sencilla basada en los artículos de lasreferencias.

```python
class LiquidNeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LiquidNeuralNetwork, self).__init__()
        self.hidden_size = hidden_size
        self.W_in = nn.Linear(input_size, hidden_size)
        self.W_hid = nn.Linear(hidden_size, hidden_size)
        self.W_out = nn.Linear(hidden_size, output_size)
        self.bias_hid = nn.Parameter(torch.zeros(hidden_size))
        self.bias_out = nn.Parameter(torch.zeros(output_size))
        self.time_constant = 0.1  # Parámetro de constante de tiempo

    def forward(self, x):
        # Estado oculto inicializado a ceros
        hidden_state = torch.zeros((x.size(0), self.hidden_size), device=x.device)
        for _ in range(1):  # Procesamos las imágenes como un solo paso
            hidden_state = (1 - self.time_constant) * hidden_state + \
                           self.time_constant * (self.W_in(x) + self.bias_hid) + \
                           torch.tanh(self.W_hid(hidden_state))
        output = self.W_out(hidden_state) + self.bias_out
        return output

```

### Función de entrenamiento

Definimos el ciclo de entrenamiento que actualizará los pesos de la red en función del error calculado.

```python
# Función de entrenamiento
def train(model, train_loader, optimizer, criterion, epochs=5):
    model.train()  # Ponemos el modelo en modo entrenamiento
    for epoch in range(epochs):
        running_loss = 0.0
        for inputs, labels in train_loader:
            optimizer.zero_grad()  # Inicializamos los gradientes
            inputs = inputs.view(inputs.size(0), -1)  # Aplanamos las imágenes
            outputs = model(inputs)  # Realizamos una pasada hacia adelante (forward)
            loss = criterion(outputs, labels)  # Calculamos la pérdida
            loss.backward()  # Retropropagamos el error
            optimizer.step()  # Actualizamos los pesos

            running_loss += loss.item()  # Acumulamos la pérdida
        print(f"Epoch {epoch+1}, Pérdida: {running_loss/len(train_loader):.4f}")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt  # For plotting

# Synthetic Time Series Dataset
class TimeSeriesDataset(Dataset):
    def __init__(self, data, targets, seq_length):
        self.data = data
        self.targets = targets
        self.seq_length = seq_length

    def __len__(self):
        return len(self.data) - self.seq_length

    def __getitem__(self, index):
        x = self.data[index:index+self.seq_length]
        y = self.targets[index+self.seq_length]
        return x, y

# Generate synthetic data
time_steps = 1000
seq_length = 10
num_features = 1  # For univariate time series

# Generate a sine wave as an example time series
t = np.linspace(0, 8 * np.pi, time_steps)
data = np.sin(t) + 0.1 * np.random.randn(time_steps)
data = data.astype(np.float32)
targets = data[1:]  # Next value prediction

# Split data into training and testing sets
train_size = int(0.8 * (len(data) - 1))
test_size = len(data) - 1 - train_size

train_data = data[:train_size + seq_length]
test_data = data[train_size:]

# Prepare training dataset
train_data_tensor = torch.from_numpy(train_data[:-1]).unsqueeze(-1)
train_targets_tensor = torch.from_numpy(train_data[1:])
train_dataset = TimeSeriesDataset(train_data_tensor, train_targets_tensor, seq_length)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Prepare testing dataset
test_data_tensor = torch.from_numpy(test_data[:-1]).unsqueeze(-1)
test_targets_tensor = torch.from_numpy(test_data[1:])
test_dataset = TimeSeriesDataset(test_data_tensor, test_targets_tensor, seq_length)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)  # Batch size 1 for plotting

print("Time series data loaded and split into training and testing sets.")
```

```python
import torch.nn as nn

class LiquidNeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LiquidNeuralNetwork, self).__init__()
        self.hidden_size = hidden_size
        self.time_constant = 0.1  # Time constant parameter

        # Layers
        self.W_in = nn.Linear(input_size, hidden_size)
        self.W_hid = nn.Linear(hidden_size, hidden_size)
        self.W_out = nn.Linear(hidden_size, output_size)
        self.bias_hid = nn.Parameter(torch.zeros(hidden_size))
        self.bias_out = nn.Parameter(torch.zeros(output_size))

    def forward(self, x):
        # x shape: (batch_size, seq_length, input_size)
        batch_size, seq_length, _ = x.size()
        hidden_state = torch.zeros((batch_size, self.hidden_size), device=x.device)

        for t in range(seq_length):
            input_t = x[:, t, :]
            hidden_state = (1 - self.time_constant) * hidden_state + \
                           self.time_constant * (self.W_in(input_t) + self.bias_hid) + \
                           torch.tanh(self.W_hid(hidden_state))
        output = self.W_out(hidden_state) + self.bias_out
        return output
```

```python
# Training Function
def train(model, train_loader, optimizer, criterion, epochs=5):
    model.train()
    for epoch in range(epochs):
        running_loss = 0.0
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)  # Forward pass
            loss = criterion(outputs.squeeze(), labels)  # Calculate loss
            loss.backward()  # Backpropagation
            optimizer.step()  # Update weights

            running_loss += loss.item()
        print(f"Epoch {epoch+1}, Loss: {running_loss/len(train_loader):.4f}")
```

```python
# Hyperparameters
input_size = num_features  # Number of features in the time series
hidden_size = 50           # You can adjust this number
output_size = 1            # Predicting a single value

# Model, Optimizer, Loss Function
model = LiquidNeuralNetwork(input_size, hidden_size, output_size)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()  # For regression tasks

# Train the model
train(model, train_loader, optimizer, criterion, epochs=10)
```

```python
# Evaluation Function
def evaluate(model, test_loader):
    model.eval()
    predictions = []
    actuals = []
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            predictions.append(outputs.item())
            actuals.append(labels.item())
    return predictions, actuals

# Get predictions and actual values
predictions, actuals = evaluate(model, test_loader)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(range(len(actuals)), actuals, label='Actual')
plt.plot(range(len(predictions)), predictions, label='Predicted')
plt.title('Actual vs Predicted Values')
plt.xlabel('Time Steps')
plt.ylabel('Value')
plt.legend()
plt.show()
```

Creamos una clase que implementa un hibrido entre LSTM y LNNs muy interesante

```python
class LSTMLiquidNeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LSTMLiquidNeuralNetwork, self).__init__()
        self.hidden_size = hidden_size
        self.time_constant = 0.1  # Parámetro que define el comportamiento temporal

        # LSTM Layer
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.W_out = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # x shape: (batch_size, seq_length, input_size)
        batch_size = x.size(0)
        hidden_state = torch.zeros(1, batch_size, self.hidden_size).to(x.device)
        cell_state = torch.zeros(1, batch_size, self.hidden_size).to(x.device)

        # LSTM forward pass
        out, (hidden_state, cell_state) = self.lstm(x, (hidden_state, cell_state))

        # Incorporar dinámica líquida en el estado oculto
        hidden_state = hidden_state * torch.exp(-torch.tensor(self.time_constant).to(x.device))

        # Get the output from the last time step
        out = self.W_out(out[:, -1, :])
        return out


```

### 4 modelos distintos

```python
# Modelo Liquid Neural Network
class LiquidNeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LiquidNeuralNetwork, self).__init__()
        self.hidden_size = hidden_size
        self.time_constant = 0.1

        self.W_in = nn.Linear(input_size, hidden_size)
        self.W_hid = nn.Linear(hidden_size, hidden_size)
        self.W_out = nn.Linear(hidden_size, output_size)
        self.bias_hid = nn.Parameter(torch.zeros(hidden_size))
        self.bias_out = nn.Parameter(torch.zeros(output_size))

    def forward(self, x):
        batch_size, seq_length, _ = x.size()
        hidden_state = torch.zeros(batch_size, self.hidden_size).to(x.device)

        for t in range(seq_length):
            input_t = x[:, t, :]
            hidden_state = (1 - self.time_constant) * hidden_state + \
                           self.time_constant * (self.W_in(input_t) + self.bias_hid) + \
                           torch.tanh(self.W_hid(hidden_state))
        output = self.W_out(hidden_state) + self.bias_out
        return output

```

```python
# Modelo LSTM
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

```

```python
# Modelo RNN Simple
class SimpleRNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(SimpleRNNModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True, nonlinearity='tanh')
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        out, _ = self.rnn(x, h0)
        out = self.fc(out[:, -1, :])
        return out
```

```python
# Modelo Liquid Neural Network con LSTM
class LiquidNeuralNetworkWithLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(LiquidNeuralNetworkWithLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.time_constant = 0.1

        # Agregamos una capa LSTM
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)

        self.W_in = nn.Linear(hidden_size, hidden_size)
        self.W_hid = nn.Linear(hidden_size, hidden_size)
        self.W_out = nn.Linear(hidden_size, output_size)
        self.bias_hid = nn.Parameter(torch.zeros(hidden_size))
        self.bias_out = nn.Parameter(torch.zeros(output_size))

    def forward(self, x):
        batch_size, seq_length, _ = x.size()
        # Inicializamos el estado oculto de la LSTM
        h_0 = torch.zeros(1, batch_size, self.hidden_size).to(x.device)
        c_0 = torch.zeros(1, batch_size, self.hidden_size).to(x.device)

        # Pasamos la entrada a través de la LSTM
        lstm_out, _ = self.lstm(x, (h_0, c_0))

        hidden_state = torch.zeros(batch_size, self.hidden_size).to(x.device)

        for t in range(seq_length):
            lstm_t = lstm_out[:, t, :]
            hidden_state = (1 - self.time_constant) * hidden_state + \
                           self.time_constant * (self.W_in(lstm_t) + self.bias_hid) + \
                           torch.tanh(self.W_hid(hidden_state))
        output = self.W_out(hidden_state) + self.bias_out
        return output
```

El entrenamiento será así

```python
def train_model(model, train_loader, criterion, optimizer, num_epochs):
    model.train()
    for epoch in range(num_epochs):
        running_loss = 0.0
        for inputs, labels in train_loader:
            inputs = inputs.to(device).float()
            labels = labels.to(device).float()

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs.squeeze(), labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.6f}")

```

```python
def evaluate_model(model, test_loader):
    model.eval()
    predictions = []
    actuals = []
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs = inputs.to(device).float()
            outputs = model(inputs)
            predictions.append(outputs.item())
            actuals.append(labels.item())
    return predictions, actuals
```

```python
# Configuración común
input_size = 1
hidden_size = 50
num_layers = 2
output_size = 1
num_epochs = 30
learning_rate = 0.001

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Entrenando en {device}")

criterion = nn.MSELoss()

```

```python
# Modelo Liquid Neural Network
model_liquid = LiquidNeuralNetwork(input_size, hidden_size, output_size).to(device)
optimizer_liquid = optim.Adam(model_liquid.parameters(), lr=learning_rate)

print("\nEntrenando Liquid Neural Network...")
train_model(model_liquid, train_loader, criterion, optimizer_liquid, num_epochs)

```

```python
predictions_liquid, actuals_liquid = evaluate_model(model_liquid, test_loader)
```

```python
# Modelo LSTM
model_lstm = LSTMModel(input_size, hidden_size, num_layers, output_size).to(device)
optimizer_lstm = optim.Adam(model_lstm.parameters(), lr=learning_rate)

print("\nEntrenando LSTM...")
train_model(model_lstm, train_loader, criterion, optimizer_lstm, num_epochs)
```

```python
predictions_lstm, actuals_lstm = evaluate_model(model_lstm, test_loader)
```

```python
# Modelo RNN Simple
model_rnn = SimpleRNNModel(input_size, hidden_size, num_layers, output_size).to(device)
optimizer_rnn = optim.Adam(model_rnn.parameters(), lr=learning_rate)

print("\nEntrenando RNN Simple...")
train_model(model_rnn, train_loader, criterion, optimizer_rnn, num_epochs)
```

```python
predictions_rnn, actuals_rnn = evaluate_model(model_rnn, test_loader)
```

```python
# Modelo Liquid Neural Network con LSTM
model_liquid_lstm = LiquidNeuralNetworkWithLSTM(input_size, hidden_size, output_size).to(device)
optimizer_liquid_lstm = optim.Adam(model_liquid_lstm.parameters(), lr=learning_rate)

print("\nEntrenando Liquid Neural Network con LSTM...")
train_model(model_liquid_lstm, train_loader, criterion, optimizer_liquid_lstm, num_epochs)

```

```python
predictions_liquid_lstm, actuals_liquid_lstm = evaluate_model(model_liquid_lstm, test_loader)
```
